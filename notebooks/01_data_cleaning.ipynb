## 2. Data Cleaning/Preprocessing

#### Handling Missing Values

```python
# Remove Players with No Games
df = df[df["g"] > 0].copy()
```

```python
df.isna().sum()
```

```python
# Checking if NA values are a MAR (missing at random) or MNAR (missing not at random)
pairs = [
    ("fg_pct",  "fga_per_g"),
    ("fg3_pct", "fg3a_per_g"),
    ("ft_pct",  "fta_per_g"),
    ("fg2_pct", "fg2a_per_g"),
]

for col, denom in pairs:
    print(f"\nChecking {col} vs {denom}")
    print(
        df.loc[df[col].isna(), denom].describe()
    )
```

```python
# shooting percentages
col = [
    "fg_pct", "fg3_pct","ft_pct", "fg2_pct"
]
df[col] = df[col].fillna(0)

# advanced stats
adv_cols = [
    "per","ts_pct","efg_pct","usg_pct",
    "orb_pct","drb_pct","trb_pct",
    "ast_pct","stl_pct","blk_pct",
    "tov_pct","ws_per_48", "fg3a_per_fga_pct", "fta_per_fga_pct"
]
df[adv_cols] = df[adv_cols].fillna(0)
```

```python
df.isna().sum()
```

#### Checking for duplicates

```python
df["player"] = df["player"].str.strip().str.lower()
```

```python
# Check for duplicate player-season-age entries
df.duplicated(subset=KEY).sum()
```

```python
#Checked for same player same season same age but on a different team
(df.groupby(KEY)["team_id"].nunique() > 1).sum()
```

#### Validating Games per Season
